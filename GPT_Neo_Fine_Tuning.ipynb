```python
from happytransformer import HappyGeneration
gpt_neo = HappyGeneration(model_type="GPT-NEO", model_name="EleutherAI/gpt-neo-125M") 

print (gpt_neo)
```

```python
from happytransformer import GENTrainArgs
```

```python
from google.colab import drive                                     #mounting the drive to acces the data stored in google drive
drive.mount('/content/drive')
```

```python
import pandas as pd
data_df = pd.read_csv('/content/drive/MyDrive/Acies/Analysis_Chart_Configs.csv')
```

```python
traine = data_df['Description']
textfile = open("/content/sample_data/train.txt", "w")
for element in traine:
    textfile.write(element + "\n")
textfile.close()
```

```python
train_args = GENTrainArgs(num_train_epochs=1, learning_rate=2e-05, batch_size=2) 

gpt_neo.train("/content/sample_data/train.txt", args=train_args)
```

```python
from happytransformer import GENSettings

top_k_sampling_settings = GENSettings(do_sample=True, top_k=50, max_length=10, min_length=5)
output_top_k_sampling = gpt_neo.generate_text("Plot Isin", args=top_k_sampling_settings)

print (output_top_k_sampling.text)
```
